## Image Display Function

```python
def render_plot(gmatrix, delta_s, start, end, energy, figsize):
    gworld = gw.GridWorld(size=[10, 10], default=True)
    custom_gworld = gworld.SetCustomGridWorld(gmatrix, delta_s, start, end)
    gworld.energy = energy
    render = gworld.RenderGridWorld(position=start)

    # Switch blue and red channels to render in matplotlib from OpenCv2 image format
    red = render[:, :, 2].copy()
    blue = render[:, :, 0].copy()
    render[:, :, 0] = red
    render[:, :, 2] = blue

    # Show image
    fig, ax = plt.subplots(nrows=1, ncols=1, figsize=figsize)
    ax.imshow(render)
```

## Gridworld Environment F (TIGHT NAVIGATION TEST)
- This environment tests the agent's capacity to overcome bottle necks in a path.
- Most often the agent can be stuck in the first and second bottle necks.
- The third "bottle neck" is the sparse reward setting and with the goal tightly placed at the end.
- If the agent overcomes the first and second bottle necks, it will have a long walk going to the sparse reward stage.
- Most often, the agent's policy can be at the brink of being overblown at the sparse reward stage. 
- This is due to the diffuclty of overcoming the two bottle necks and the long walk.

**Optimal score = 7**
